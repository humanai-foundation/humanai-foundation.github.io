# Library

```python
import pandas as pd
```

# Read Files

```python
df_boards = pd.read_csv('../../Datasets/CleanedData/boards_cleaned.csv')
df_members = pd.read_csv('../../Datasets/CleanedData/members_cleaned.csv')
df_posts = pd.read_csv('../../Datasets/CleanedData/posts_cleaned.zip')
df_threads = pd.read_csv('../../Datasets/CleanedData/threads_cleaned.csv')
```

```python
df_boards.shape, df_members.shape, df_posts.shape, df_threads.shape
```

```python
df_boards_threads = pd.merge(df_boards, df_threads, left_on='id', right_on='board_id', suffixes=('_board', '_thread')).drop(columns='board_id').rename(columns={'creator': 'creator_thread', 'creator_id': 'creator_id_thread'})
print(df_boards_threads.shape)
df_boards_threads.head()
```

```python
df_boards_threads_members = pd.merge(df_boards_threads, df_members, left_on=['creator_id_thread', 'creator_thread'], right_on=['id', 'username'], suffixes=('_thread', '_member')).drop(columns=['id', 'username'])
print(df_boards_threads_members.shape)
df_boards_threads_members.head()
```

```python
df_final = pd.merge(df_boards_threads_members, df_posts, left_on=['id_thread', 'id_board'], right_on=['thread_id', 'board_id'], suffixes=('', '_post')).drop(columns=['id', 'id_thread', 'id_board', 'thread_id', 'board_id']).rename(columns={'creator': 'creator_post', 'creator_id': 'creator_id_post'})
print(df_final.shape)
df_final.head()
```

# Preprocess Df-Final

```python
df_final.drop_duplicates(inplace=True)
df_final.dropna(subset='content', inplace=True)
print(df_final.shape)
df_final.head()
```

```python
df_final.to_csv('merged_data.zip', index=False)
```
